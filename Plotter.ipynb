```julia
using Knet
using JLD
using DelimitedFiles
using Plots; default(fmt=:png,ls=:auto)
```

```julia
history = Knet.load("History3(2).jld2", "history");
loss_tot = []
trn_acc = []
tst_acc =[]
for i=1:length(history)
    temp_losstot = history[i][1]
    temp_trnacc = history[i][5]
    temp_tstacc = temp_tracc =history[i][10] 
    push!(loss_tot, temp_losstot)
    push!(trn_acc, temp_trnacc)
    push!(tst_acc, temp_tstacc)    
end
plot(loss_tot, label = "Total Loss", title=" Momentum Optimizer lr = 0.003, gamma = 0.95")
```

```julia
plot([[trn_acc],[tst_acc]], title = "Momentum Optimizer lr = 0.003, gamma = 0.95", label=[:Train_Acc, :Test_Acc])
```

```julia
writedlm("Acc.csv", (trn_acc, tst_acc), ',')
```

```julia
plot(tst_acc, label = "Test Accuracy", title=" Momentum Optimizer lr = 0.003, gamma = 0.95")
```

## Batch Norm with Detached BNorm weights

```julia
elt = Knet.load("HistoryBnorm.jld2");
a = elt["(history, loss_batch)"]
history = a[1]
loss_batch = a[2];

loss_tot = []
trn_acc = []
tst_acc =[]
for i=1:length(history)
    temp_losstot = history[i][1]
    temp_trnacc = 100 .* history[i][5]
    temp_tstacc = 100 .* history[i][10] 
    push!(loss_tot, temp_losstot)
    push!(trn_acc, temp_trnacc)
    push!(tst_acc, temp_tstacc)    
end
plot(loss_tot, label = "Total Loss", title=" Adam = 0.003")
```
